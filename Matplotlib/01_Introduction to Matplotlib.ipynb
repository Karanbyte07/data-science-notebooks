```python
#Sachin Tendulkars Runs Over Time
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

#adding x and y labels
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")

plt.plot(years, runs)
```

#### Multiple lines in one plot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label = "Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.legend()  #different data set ko alag alg colour me show krna
plt.xlabel("Year") #x axis ka name
plt.ylabel("Runs Scored") #y axis ka name
plt.title("Performance Comparison") #title of the graph
```

#### Format strings

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
#anothere way to represent for readbility
plt.plot(years, kohli, color='pink', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

#### Give Thickness and easy visualization no overlap anything 

```python
# Plot Kohli's data against years 
plt.plot(years, kohli, 
         color='skyblue',     # Line color is sky blue
         linestyle='--',      # Dashed line style
         linewidth=3,         # Line thickness = 3
         label="Karan")       # Label for legend

# Plot Sehwag's data against years
plt.plot(years, sehwag, 
         color='red',         # Line color is red
         linestyle='-.',      # Dash-dot line style
         linewidth=2,         # Line thickness = 2
         label="Jeet")        # Label for legend

plt.legend()       # Show legend box with labels "Karan" and "Jeet"
plt.grid()         # Add grid lines in background for better readability
plt.tight_layout() # Adjust spacing so labels/legend dont overlap
```
