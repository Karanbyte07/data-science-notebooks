```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs) #in x axis years, and y-axis runs
plt.xlabel("Years") #label in x axis is years
plt.ylabel("Runs") #label in y axis is runs
plt.title("Sachin Tendulkar's Score") #Title of the bar 
plt.show() #to show the bar
```

#### Let’s compare Sachin, Sehwag, and Kohli side-by-side for the same years.

```python
import numpy as np
 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years)) #index positions
width = 0.25
```

- x - width / x / x + width → x‑axis positions (shift left, center, or right for side‑by‑side bars) so no overlap
- sachin / sehwag / kohli → heights of bars (data values = runs)
- width=width → thickness of each bar
- label="..." → name shown in legend
- color="..." → bar color for visual distinction


```python
# Plotting side-by-side bar charts for Sachin, Sehwag, and Kohli

# Plot Sachin's runs shifted slightly to the left of the x-axis positions
plt.bar(x - width, sachin, width=width, label="Sachin", color="orange")

# Plot Sehwag's runs centered at the x-axis positions
plt.bar(x, sehwag, width=width, label="Sehwag", color="yellow")

# Plot Kohli's runs shifted slightly to the right of the x-axis positions
plt.bar(x + width, kohli, width=width, label="Kohli", color="Green")

# Label for the x-axis
plt.xlabel("Years")

# Label for the y-axis
plt.ylabel("Runs")

# Title of the chart
plt.title("Run Comparison")

# Set custom tick labels on the x-axis (years) in bottom 
plt.xticks(x, years)

# Show legend to differentiate players
plt.legend()

# Adjust layout to prevent overlapping of labels and elements
plt.tight_layout()
```

#### Horizontal bars

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs)
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
```

Adding Value Labels with plt.text()    

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
 
plt.bar(players, runs, color="skyblue")

#add labels in top of the bar
for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha = 'center')
# i - This is the x-coordinate. Likely represents the index of a data point (like the position of a bar in a bar chart).
# runs[i] + 50 - This is the y-coordinate. It places the text slightly above the bar (or point) by adding 50 units to the value runs[i].

#- ha='center' ensures the text is horizontally centered above the bar.

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
```
