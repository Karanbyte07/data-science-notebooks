```python
import matplotlib.pyplot as plt

students = ["A","B","C","D","E","F","G","H","I","J"]
marks = [45, 78, 56, 89, 90, 67, 55, 76, 88, 92]
```

#### Category Comparison - Bar chart

```python
plt.bar(students, marks)
plt.title("Marks Of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
```

#### Pie Chart → Percentage Distribution

```python
plt.pie(marks, labels=students, autopct='%1.1f%%')
plt.title("Marks Distribution")
plt.legend(loc = "upper right", bbox_to_anchor=(1.2, 1))
plt.tight_layout()
```

#### Histogram → Distribution of Marks

```python
plt.hist(marks, bins=5, color='skyblue', edgecolor = 'black')
plt.title("Marks Distribution Histogram")
```

#### Scatter Plot → Relationship

```python
hours_study = [2,4,3,6,7,5,3,4,6,8]
plt.scatter(hours_study, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
```

#### Subplots → Multiple Views Together

```python
fig, axs = plt.subplots(1,2, figsize=(10,4))
axs[0].bar(students, marks)
axs[0].set_title("Bar Plot")
axs[1].hist(marks, bins=5)
axs[1].set_title("Histogram")
```

#### Seaborn Example → Stylish Visualization

```python
import seaborn as sns
sns.boxplot(marks)
plt.title("Box Plot of Marks")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Demo dataset
students = ["A","B","C","D","E","F","G","H","I","J"]
marks = [45, 78, 56, 89, 90, 67, 55, 76, 88, 92]
hours_study = [2,4,3,6,7,5,3,4,6,8]

# Create subplots grid
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# 1. Bar Plot
axs[0,0].bar(students, marks, color='skyblue')
axs[0,0].set_title("Bar Plot: Marks of Students")

# 2. Pie Chart
axs[0,1].pie(marks, labels=students, autopct='%1.1f%%')
axs[0,1].set_title("Pie Chart: Marks Distribution")

# 3. Histogram
axs[1,0].hist(marks, bins=5, color='lightgreen', edgecolor='black')
axs[1,0].set_title("Histogram: Marks Spread")

# 4. Scatter Plot
axs[1,1].scatter(hours_study, marks, color='red')
axs[1,1].set_title("Scatter Plot: Study Hours vs Marks")
axs[1,1].set_xlabel("Study Hours")
axs[1,1].set_ylabel("Marks")

# 5. Line Plot (trend)
axs[2,0].plot(students, marks, marker='o', color='purple')
axs[2,0].set_title("Line Plot: Marks Trend")

# 6. Seaborn Box Plot
sns.boxplot(marks, ax=axs[2,1], color='orange')
axs[2,1].set_title("Box Plot: Marks Distribution")

plt.tight_layout()
plt.show()
```
