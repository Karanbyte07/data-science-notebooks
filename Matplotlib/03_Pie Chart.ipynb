```python
import matplotlib.pyplot as plt

plt.style.use("ggplot") 
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career runs of Indian Batsman")
plt.pie(runs, labels = labels)
```

```python
#custom colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels = labels, colors = colors)
```

```python
#Add Edges and Style Slices with wedgeprops
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
```

```python
#Highlight a Slice Using explode
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice
 
plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
```

```python
#Add Shadows for a 3D Feel
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
```

```python
#show percentage
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
```

```python
#merging them final pie chart
Expenses = ["rent", "food", "travel", "Entertainment"]
Amounts = [12000, 5000, 3000, 2000]

plt.pie(
    Amounts, 
    labels = Expenses,              # Labels for each slice
    autopct = '%1.0f%%',            # Show percentage values (no decimals)
    colors = ["#a6cee3", "#fb9a99", "#b2df8a", "#fdbf6f"],  # Pastel color palette
    wedgeprops = {"edgecolor":"white", "linewidth":2},      # White borders for clean separation
    explode = [0, 0, 0, 0.1],       # Pop out 'Entertainment' slice slightly
    startangle = 90                 # Rotate chart to start at 12 oclock
    # shadow = True                  # Optional: add shadow for depth
)

# Title styling
plt.title("Monthly Expenses Distribution", fontsize=14, fontweight="bold")

# Legend for clarity
plt.legend(Expenses, title="Categories", loc="upper right", bbox_to_anchor=(1.5, 1))
plt.tight_layout()
```

- labels  The text shown in the legend (often same as your slice labels).
- title  Adds a title above the legend box.
- loc  Position of the legend ("upper right", "lower left", "center", etc.).
- bbox_to_anchor  Finetunes the legends position outside the plot area.
- fontsize  Controls the size of the legend text.
- frameon  True/False to show or hide the legend box border.
- ncol  Number of columns in the legend (useful if you want items sidebyside).
