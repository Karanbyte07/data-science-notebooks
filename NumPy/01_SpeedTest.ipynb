In this section we will check which is faster numpy processing or simple python 

```python
#To zip the two lists 

l1 = [2,4,6,8]
l2 = [3,1,5,7]

ans = list(zip(l1,l2)) #zip object return krta hai toh usko typecast ke liye list me kiya hamne 
print(ans) #jo ki tuples dega output me
```

```python
#Using Python lists
import time

size = 1_000_000 #a very big size of data list

l1 = list(range(size))
l2 = list(range(size))

start =  time.time() #starting time 
add = [x + y for x,y in zip(l1,l2)]
end = time.time() #ending time

print(end - start) #total time taken
print(add[0:10])
```

```python
#Using numpy lists
import numpy as np
import time

size = 1_000_000 #a very big size of data list

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start =  time.time() #starting time 
add = l1+l2
end = time.time() #ending time

print(end - start) #total time taken
print(add[0:10])
```

```python
#checking time speed of squaring of a number
#using python  list (loop based)

import time

list1 = [2,3,4]
start = time.time()
square = [x ** 2 for x in list1]
end = time.time()

print(f"the time taken to perform this operation: {start-end}")
print(square)
```

```python
#using numpy
import time
import numpy

arr = np.array([2,3,4])
start = time.time()
square = arr ** 2
end = time.time()

print(f"the time taken to perform this operation: {start-end}")
print(square)
```
