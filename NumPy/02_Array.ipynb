```python
#creating numpy array
import numpy as np

arr1 = np.array([1,2,3,4]) #array with list
print(f" 1D array: {arr1} ")

arr2 = np.array([[1,2,3], [7,8,4]])
print(f" 2D array: {arr2} ")

print(f"type of arr1 -> {type(arr1)}")
print(f"type of arr2 -> {type(arr2)}")
print(f"shape of arr1 -> {arr1.shape}")
```

```python
#create array with elements 0
arr1 = np.zeros((3,3))
print (arr1)
```

```python
#array with elements 1 with 2 row 3 columns
np.ones((2,3))
```

```python
#array with element full of (number)
arr3 = np.full((3,2), 5)
print(arr3)
```

```python
#identity matrix of size (num)
np.eye(3)
```

```python
#in range we want array with adding number of 2
np.arange(1,10, 2) # o/p 1+2, 3+2,..
```

```python
#equi spaced value in range of(2,10) , total elemnts with 5
np.linspace(2,10, 5)
```

## Array Properties

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])
```

```python
arr
```

```python
print("Shape:", arr.shape)   # (2, 3) → 2 rows, 3 columns
print("Size:", arr.size)     # 6 → total elements
print("Dimensions:", arr.ndim) # 2 → 2D array
print("Data type:", arr.dtype) # int64 (or int32 on Windows)
```

## Changing Data Types

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70,80,90], [2,3,5]], dtype=np.float64)  # Explicit type)
```

```python
arr.dtype
```
