```python
import numpy as np
arr = np.array([2,4,6,8])
arr.dtype #to know what is datatype in numpy
```

```python
arr.astype("float64") # convert arr to float (changing data type using astype)
```

```python
arr2 = np.array([10,20,30,40,50], dtype = "float64") #cast the datatype
arr2 
```

```python
arr64 = np.array([1,2,3], dtype = np.int64)
arr32 = np.array([4,5,6], dtype = np.int32)
print(arr64.nbytes) #print the scpace ocuupied of array
print(arr32.nbytes)
```
