## Match case is similar to switch statements

```python
https_request = 600
match https_request:
    case 200:
        print("Response OK")
    case 404:
        print("Not Found")
    case 400:
        print("Bad Response")
    case 500:
        print("Internal Server Error")
    case 502:
        print("Bad Gateway")
    case _: 
        print("Nothing matched")
print("End of Program")
```

```python
point = (3,0)
match point:
    case (0,0):
        print("origin")
    case(x,0):
        print(f"point at {x} axis")
    case(0,y):
        print(f"point at {y} axis")
    case(x,y):
        print(f"point at ({x} , {y})")
```
