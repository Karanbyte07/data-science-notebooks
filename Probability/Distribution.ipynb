```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Continuous uniform from 0 to 1
samples = np.random.uniform(0, 1, 10000)
## o aur 1 ke beich meh random 10000 number genrate kre

plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue')
plt.title("Continuous Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import binom
#probabilty of exactly 6 heads in 10 tosses of coin (p = 0.5)
prob = binom.pmf(k = 6, n = 10, p = 0.5)
print(prob)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate values
x = np.linspace(-4, 4, 1000)
mean = 0
std_dev = 1

# Get the probability density
y = norm.pdf(x, loc=mean, scale=std_dev)

# Plot
plt.plot(x, y)
plt.title("Standard Normal Distribution (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
data = np.random.normal(loc=70, scale=20, size=2000)
plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
