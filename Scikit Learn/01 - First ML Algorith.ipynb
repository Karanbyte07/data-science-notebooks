```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [
    [150,1],
    [170, 0],
    [130,1],
    [120,0]
]
```

```python
labels = ["orange", "apple", "orange", "apple"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[160,0]])
```

```python
print(f"The fruit is {prediction[0]}")
```

```python
#now visualize this by using making tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plot_tree(clf, feature_names = ["weight", "color"], class_names=["orange", "apple"], filled=True)
```
