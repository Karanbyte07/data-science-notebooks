We are given a dataset of users, where each user has:
- A unique `id`
- A list of `friends`
- A list of `liked_pages`

We want to **recommend new pages** to a given user based on the pages liked by other users.  
The intuition:  
- If two users share some liked pages, then the other userâ€™s liked pages (which the current user hasnâ€™t liked yet) are good candidates for recommendation.  
- Pages that appear more often in such overlaps should be ranked higher.

# âš™ï¸ Page Recommendation Algorithm

## Steps
1. **Load the data**
   - Read the JSON file into a Python dictionary.

2. **Build a mapping of users â†’ liked pages**
   - For each user, store their liked pages in a dictionary for quick lookup.

3. **Check if the target user exists**
   - If not, return an empty list.

4. **Get the target userâ€™s liked pages**
   - Store them in a set for fast intersection operations.

5. **Initialize a score dictionary for page suggestions**
   - This will hold pages as keys and their recommendation score as values.

6. **Loop through all other users**
   - Skip the target user.
   - Find the intersection of liked pages between the target user and the other user.
   - For each page liked by the other user:
     - Increase its score by the number of shared pages.

7. **Remove pages already liked by the target user**
   - We donâ€™t want to recommend pages they already like.

8. **Sort the pages by score (descending)**
   - Higher score = stronger recommendation.

9. **Return the sorted list of page IDs**
   - This is the recommendation output.

````python
import json


def load_data(filename):
    with open(filename, "r") as f:
        return json.load(f)


def pages_you_might_like(user_id, data):
    user_pages = {}
    for user in data["users"]:
        user_pages[user["id"]] = set(user["liked_pages"])

    # if the user is not found, return an empty list
    if user_id not in user_pages:
        return []

    user_liked_pages = user_pages[user_id]
    page_suggestion = {}

    for other_user, pages in user_pages.items():
        if other_user != user_id:
            shared_pages = user_liked_pages.intersection(pages)
            for page in pages:
                # increment score based on shared pages
                page_suggestion[page] = page_suggestion.get(page, 0) + len(shared_pages)

    sorted_pages = sorted(page_suggestion.items(), key=lambda x: x[1], reverse=True)
    return [page_id for page_id, _ in sorted_pages]


# load the data
data = load_data("massive_data.json")
user_id = 1
page_recommendation = pages_you_might_like(user_id, data)
print(page_recommendation)
````
